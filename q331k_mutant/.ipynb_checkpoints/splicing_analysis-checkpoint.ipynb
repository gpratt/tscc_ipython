```python
def larger_and_smaller(genes, values):
    max_idx = uniquely_mapped_reads.ix[genes].idxmax()
    min_idx = uniquely_mapped_reads.ix[genes].idxmin()
    return max_idx, min_idx
```

```python
#less Stringent method
results = {}
total_counts = {}
for genotype_1, genotype_2 in interesting_combinations:

    combination = unstressed_splicing_df[(unstressed_splicing_df.rep1_genotype == genotype_1) & (unstressed_splicing_df.rep2_genotype == genotype_2)]
    total_counts[(genotype_1, genotype_2)] = combination.number_of_sites.values
    rep1_max, rep1_min = larger_and_smaller(set(combination.index.get_level_values(level=0)), uniquely_mapped_reads)
    rep2_max, rep2_min = larger_and_smaller(set(combination.index.get_level_values(level=1)), uniquely_mapped_reads)

    
    rep1, rep2, rep3, rep4 = combination[0]
    
    rep1 = combination.ix[rep1_max, rep2_min][0]
    rep2 = combination.ix[rep1_min, rep2_max][0]

    rep1 = pd.read_table(rep1, index_col=0)
    rep2 = pd.read_table(rep2, index_col=0)
    result = pd.concat({"rep1": rep1, "rep2": rep2}, names=['rep', 'event'])

    results[(genotype_1, genotype_2)] = result
interesting_splicing_df = pd.concat(results)
interesting_splicing_df['gene_id'] = [event_to_gene_id[event] for event in interesting_splicing_df.index.get_level_values(level="event")]
```

```python
diff_threshold = .1
bayes_factor_threshold = 1

def sig_event(event, diff=.1, bayes_factor = 1):
    return np.abs(event['diff']) > diff and event['bayes_factor'] > bayes_factor
```

```python
#Stringent filtering
# results = {}
# merged_df = {}
# for genotype_name, df in interesting_splicing_df.groupby(level=[0,1]):
#     counts = Counter()
#     for x, (name, df) in enumerate(df.groupby(level="event")):
#         counts[len(df)] += 1
#         if len(df) >= 2 and any(df.bayes_factor > 1) and all(np.abs(df['diff']) > .1) and (all(df['diff'] > 0) or all(df['diff'] < 0)):
#             counts['confident'] += 1
#             merged_df[tuple(list(genotype_name) + [name])] = df.mean()
#         elif len(df) >= 2 and all(df.bayes_factor > 1) and all(np.abs(df['diff']) > .1) and not (all(df['diff'] > 0) or all(df['diff'] < 0)):
#             counts['discordent'] += 1

#         counts["diff_" + str(sum(np.abs(df['diff']) > .1))] += 1
#         counts["bayes_factor_" + str(sum(np.abs(df['bayes_factor']) > .1))] += 1
#     results[genotype_name] = counts
```

```python
#less stringent filtering
results = {}
merged_df = {}
for genotype_name, df in interesting_splicing_df.groupby(level=[0,1]):
    counts = Counter()
    for x, (name, df) in enumerate(df.groupby(level="event")):
        counts[len(df)] += 1
        #The last arg is an xor both events must be changing in the same direction
        if len(df) >= 2:
            one, two = df['diff'] > 0
            if (sig_event(df.irow(0), diff_threshold, bayes_factor_threshold) or sig_event(df.irow(1), diff_threshold, bayes_factor_threshold)) and (not one ^ two):
                counts['confident'] += 1
                merged_df[tuple(list(genotype_name) + [name])] = df.mean()
        elif len(df) >= 2 and all(df.bayes_factor > bayes_factor_threshold) and all(np.abs(df['diff']) > diff_threshold) and not (all(df['diff'] > 0) or all(df['diff'] < 0)):
            counts['discordent'] += 1

        counts["diff_" + str(sum(np.abs(df['diff']) > diff_threshold))] += 1
        counts["bayes_factor_" + str(sum(np.abs(df['bayes_factor']) > bayes_factor_threshold))] += 1
    results[genotype_name] = counts
```

```python
def plot_both_changing(genotype_1, vs_1, genotype_2, vs_2, ax):
    significantly_enriched_genes_either = significant_df.ix[genotype_1, vs_1].index.get_level_values(level=0) | significant_df.ix[genotype_2, vs_2].index.get_level_values(level=0)
    foo = unfiltered_interesting_splicing_df[unfiltered_interesting_splicing_df.event.isin(significantly_enriched_genes_either)]
    changing_in_either = foo['diff'].unstack()
    print len(significantly_enriched_genes_either)

    KA, WT, middle = ['r','b', 'g']
    ka_genes = significant_df.ix[genotype_1, vs_1].index.get_level_values(level=0)
    wt_genes = significant_df.ix[genotype_2, vs_2].index.get_level_values(level=0)


    colors = []
    for gene in changing_in_either.columns:
        if gene in ka_genes and gene in wt_genes:
            colors.append(middle)
        elif gene in ka_genes:
            colors.append(KA)
        elif gene in wt_genes:
            colors.append(WT)

    ax.scatter(changing_in_either.ix[genotype_1, vs_1].values, changing_in_either.ix[genotype_2, vs_2].values, alpha=.7, 
               color=colors, 
               s=5)
    ax.set_xlabel("{} vs {} $\Delta \Psi$".format(names[genotype_1], names[vs_1]), fontsize=10)
    ax.set_ylabel("{} vs {} $\Delta \Psi$".format(names[genotype_2], names[vs_2]), fontsize=10)
    #ax.set_title("Differental Espression of WT vs KA Mutant")
    ax.set_xlim(-1, 1)
    ax.set_ylim(-1, 1)

    red_patch = mpatches.Patch(color='r', label="{} vs {}".format(names[genotype_1], names[vs_1]))
    blue_patch = mpatches.Patch(color='b', label="{} vs {}".format(names[genotype_2], names[vs_2]))
    green_patch = mpatches.Patch(color='g', label='Both')


    ax.legend(handles=[red_patch, blue_patch, green_patch], loc=0)
    sns.despine(ax=ax)
    changing_in_either = changing_in_either.dropna(axis=1)
    print stats.pearsonr(changing_in_either.ix[genotype_1, vs_1].values, changing_in_either.ix[genotype_2, vs_2].values)
    return changing_in_either
    return foo
```

```python
mm10GO = GO.GO(GOFile="/projects/ps-yeolab/genomes/mm10/mm10_ENSEMBL_TO_GO.txt.gz")
```

```python
#Factor this to a helpers function
def gencode_to_ensembl(gene_list):
    for gene in gene_list:
        yield gene.split(",")[0]
```

```python
go_results = {}
for name, df in included_significant_df.groupby(level=[0,1]):
    go = mm10GO.enrichment(gencode_to_ensembl(df.gene_id.dropna()))
    go = go[go['Bonferroni-corrected Hypergeometric p-Value'] < .05]
    go_results[tuple(['included'] + list(name))] = go
    
for name, df in excluded_significant_df.groupby(level=[0,1]):
    go = mm10GO.enrichment(gencode_to_ensembl(df.gene_id.dropna()))
    go = go[go['Bonferroni-corrected Hypergeometric p-Value'] < .05]
    go_results[tuple(['excluded'] + list(name))] = go

go_results = pd.concat(go_results, names=['direction', 'genotype_1', 'genotype_2', 'GO Term ID'])
```

```python
new_index = []
for row in go_results.index:
    inc, gt_1, gt_2, go_term = row
    new_index.append([inc, names[gt_1], names[gt_2], go_term])
go_results.index = pd.MultiIndex.from_tuples(new_index)
```
