```python
import pandas as pd
import os

from gscripts.encode import encode_helpers
import pandas as pd
```

```python
encff_files = pd.read_excel("/home/gpratt/Dropbox/EricGabe_ENCODE/eCLIP_review_120716.xlsx", "md5sum3.tsv")
encff_files = encff_files[encff_files['submitted file name'].apply(lambda x: x.endswith(".fastq.gz"))]
encff_files['submitted_files'] = encff_files['submitted file name'].apply(lambda x: os.path.basename(x).split(".")[0])
```

```python
manifest = encode_helpers.get_lab_manifest()
manifest['out_file_name'] = manifest.ENCODE_ID + "_" + manifest.RBP
```

```python
manifest
```

```python
merged_data = pd.merge(encff_files, manifest, left_on='submitted_files', right_on='out_file_name', how='left')
```

```python
merged_data = merged_data.rename(columns={'file accession': 'ENCFF_file', 
                    'inline_1': "barcode_1", 
                    'inline_2': "barcode_2",
                    'randomer_length': "UMI"})
```

```python
merged_data_filtered = merged_data[['ENCFF_file', 'barcode_1', 'barcode_2', 'UMI']]
merged_data_filtered.to_csv("/home/gpratt/Dropbox/EricGabe_ENCODE/ENCFF_barcodes.tsv", index=False, sep="\t")
```
