```python
import pandas as pd
from itertools import zip
```

```python
all_files = pd.read_table("lng_file_sizes.txt", sep=" ", header=None, names=['file_size', 'fn'])
all_files['barcode'] = all_files.fn.apply(lambda x: x.split(".")[1].split("_")[0])
all_files['exp'] = all_files.fn.apply(lambda x: x.split(".")[0])
all_files.index = pd.MultiIndex.from_tuples(zip(all_files.exp, all_files.barcode), names=["exp", "barcode"])
all_files = all_files.drop(["fn", 'barcode', 'exp'], axis=1)
all_files = all_files.unstack().T
result = pd.concat({"counts": all_files, "fraction": all_files / all_files.sum()})
result.to_csv("/home/gpratt/Dropbox/encode_integration/for_eric/LNG_errors.csv")
```
