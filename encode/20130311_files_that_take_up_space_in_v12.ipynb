```python
import pandas as pd
from IPython.core.display import HTML
```

```python
def commas(x):
    try:
        return "{:,}".format(float(x))
    except:
        return "0"
```

```python
table = pd.read_table("/home/gpratt/file_sizes.txt", header=None, names=["size", "file"])

new_col = []
for x, (name, item) in enumerate(table['file'].iteritems()):
    result = ".".join(item.split(".")[3:])
    if len(result) == 0:
        result = "fastqc"
    new_col.append(result)
 
table['file_type'] = new_col

total_contribution = table.groupby("file_type").sum()
fraction_contribution = total_contribution / total_contribution.sum() 

HTML(fraction_contribution.sort_values("size", ascending=False).to_html(formatters={"size": commas}))
```
