```python
def eplouge(count):
    return """#!/bin/bash
#PBS -N barcode_spliter
#PBS -l nodes=1:ppn=1
#PBS -o 20150224_splitter.out
#PBS -e 20150224_splitter.err
#PBS -V
#PBS -q home-yeo
#PBS -W group_list=yeo-group
#PBS -t 1-{}
#PBS -l walltime=12:00:00
cd /home/gpratt/projects/encode/data/encode_split
echo "hello, starting"
""".format(count)

prolouge = "eval ${cmd[$PBS_ARRAYID]}"
```

```python
def format_randomer_script(fn, in_dir="/home/gpratt/projects/encode/analysis/encode_split/", out_dir="/home/gpratt/projects/encode/data/van_nostrand_geo_submission"):
    in_file = os.path.join(in_dir, fn) 
    out_file = os.path.join(out_dir, ".".join(fn.split(".")[:2]) + ".randomer.fastq.gz")

    return "python /home/gpratt/gscripts/gscripts/clipseq/add_randomer.py -i {} -o {}".format(in_file, out_file)
```
