```python
def check_order(fn):
    count = 0
    with open(fn + ".mate1") as mate1:
        with open(fn + ".mate2") as mate2:
            while True:
                try:
                    count += 1
                    r1_id = mate1.next()
                    r1_seq = mate1.next()
                    r1_null = mate1.next()
                    r1_qscore = mate1.next()
                    r2_id = mate2.next()
                    r2_seq = mate2.next()
                    r2_null = mate2.next()
                    r2_qscore = mate2.next()
                except StopIteration:
                    break
                if r1_id.split("/")[0] != r2_id.split("/")[0]:
                    print "out of order"
                    break
    print count
            
```

```python
def check_order(r1, r2, outfile):
    count = 0
    with open(r1) as mate1:
        with open(r1) as mate2:
            while True:
                try:
                    count += 1
                    r1_id = mate1.next()
                    r1_seq = mate1.next()
                    r1_null = mate1.next()
                    r1_qscore = mate1.next()
                    r2_id = mate2.next()
                    r2_seq = mate2.next()
                    r2_null = mate2.next()
                    r2_qscore = mate2.next()
                except StopIteration:
                    break
                if r1_id.split("/")[0] != r2_id.split("/")[0]:
                    return "out of order"
                    
    return "good"
```

```python
def get_read_pairs(samfile):
    while True:
        try:
            r1 = samfile.next()
            r2 = samfile.next()
            
            if r1.query_name != r2.query_name:
                r1 = r2
                r2 = samfile.next()
            yield r1, r2
        except:
            break
```
